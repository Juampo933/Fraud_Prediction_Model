- Casos de uso de la delincuencia financiera:

Toma de posesi√≥n de la cuenta: La toma de posesi√≥n de la cuenta se produce cuando un estafador utiliza credenciales robadas para acceder a una cuenta en l√≠nea existente, por ejemplo, en un banco o comerciante.

Sim Swapping: Se trata de una forma de ATO en la que el estafador utiliza la informaci√≥n personal de la v√≠ctima, robada de una filtraci√≥n de datos o extra√≠da de otras fuentes de informaci√≥n, como las redes sociales, para convencer a la compa√±√≠a de telefon√≠a m√≥vil de que transfiera el n√∫mero de tel√©fono de la v√≠ctima al tel√©fono m√≥vil del estafador.

Phishing: Un ataque de phishing es cuando el estafador se hace pasar por un sitio web leg√≠timo en un correo electr√≥nico o un texto para conseguir que la v√≠ctima finalmente divulgue informaci√≥n personal o transfiera fondos.

Mientras que la IA es la implementaci√≥n inform√°tica de un pensamiento o proceso de toma de decisiones similar al humano, el aprendizaje autom√°tico utiliza t√©cnicas de algoritmos matem√°ticos para extraer relaciones complejas dentro de los datos analizados. El an√°lisis del fraude utiliza el aprendizaje autom√°tico para examinar todos los datos pertinentes relativos a una transacci√≥n y asigna una puntuaci√≥n de riesgo a la misma. Con el aprendizaje autom√°tico supervisado, el modelo se entrena utilizando datos etiquetados (datos de fraude y otros datos) y predice la probabilidad de fraude. Se entrena un modelo supervisado present√°ndole tanto eventos fraudulentos como leg√≠timos y ejecut√°ndolo para desarrollar un conjunto de instrucciones o algoritmo que se aplica a otros ejemplos. El modelo entrenado puede entonces identificar patrones tanto desconocidos como conocidos para producir una puntuaci√≥n de riesgo precisa para una transacci√≥n solicitada.

# **M√©tricas**

image.png

Recordemos que el Accuracy es el porcentaje de aciertos de mi Modelo sobre el Total de Observaciones.

Es decir, los VP, aquellas operaciones que el modelo predijo que ser√≠an fraude y lo fueron efectivamente; y los VN, aquellos que el modelo predijo que no ser√≠an fraude y no lo fueron; sobre el total de transacciones.


***Es un buen indicador a nivel general, pero no debemos quedarnos s√≥lo con √©l.***

**! Precisi√≥n:**

En cambio, la Precision (de la clase 1 en este caso) hace referencia a que tan confiable es el modelo en asignar una operacion a una clase o a otra.

Es decir, que representa el total de aciertos de casos positivos (predicciones positivas correctas sobre fraudes) sobre todos los casos donde ***el modelo predijo*** que era fraude (predicciones positivas correctas e incorrectas sobre los fraudes)

*En este caso estamos dandole m√°s importancia a los Falsos Positivos, queremos reducirlos al m√°ximo.*

*Supongamos para este caso, que la empresa no quiera que sus clientes se vean frustrados al no poder hacer transacciones correctamente cuando el modelo detecta un mont√≥n de casos de fraude cuando no lo son.*



! **Recall:**

**Asimismo, el Recall (de la clase 1) expresa cuan bien puede el modelo detectar a esa clase.**



Es decir, que representa el total de aciertos de casos positivos (predicciones positivas correctas sobre fraudes) sobre ***todos los casos donde realmente hubo fraude*** (predicciones positivas correctas y negativas incorrectas sobre los fraudes)

En este caso estamos dandole m√°s importancia a los Falsos Negativos, queremos reducirlos al m√°ximo posible.

En la Deteccion de Fraudes cobra vital importancia los Falsos Negativos.

**De nuevo, son operaciones que el sistema clasific√≥ como normales pero eran fraudulentas.**

*En este caso, es entendible que la empresa no va a utilizar un modelo en donde las operaciones que sean fraudulentas "se filtren" entre las normales,  perjudicando a los clientes y a la operatoria de la misma.*



---



**Por lo tanto, en este caso, debemos priorizar el Recall o Recuperaci√≥n como m√©trica para medir la calidad de nuestro clasificador en general.**

---



Adem√°s:

Tenemos cuatro casos posibles para cada clase:

Alta precision y alto recall: el modelo maneja perfectamente esa clase

Alta precision y bajo recall: el modelo no detecta la clase muy bien, pero cuando lo hace es altamente confiable.

Baja precisi√≥n y alto recall: La clase detecta bien la clase pero tambi√©n incluye muestras de otras clases.

Baja precisi√≥n y bajo recall: El modelo no logra clasificar la clase correctamente.


***Cuando tenemos un dataset con desequilibrio, suele ocurrir que obtenemos un alto valor de precisi√≥n en la clase Mayoritaria y un bajo recall en la clase Minoritaria***
